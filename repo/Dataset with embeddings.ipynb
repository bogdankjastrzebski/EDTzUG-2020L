# Konwersja czasowników na embeddingi

```python
verbs = list(map(lambda x: x.split(",")[0], finalfinal.verb.values))
```

```python
def to_emb(verb):
    s = Sentence(verb)
    epl.embed(s)
    emb = torch.zeros(300)
    for token in s:
        emb += token.embedding.to(torch.device('cpu'))

    return emb.data.numpy()
```

```python
embeddings = np.r_[list(map(to_emb, verbs))]
```

```python
embeddings = pd.DataFrame(embeddings)
```

```python
finalfinalfinal = pd.concat([embeddings, finalfinal.reset_index()], axis=1)
```
